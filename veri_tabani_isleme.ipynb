```python
# Gerekli Kütüphaneleri Yükleyin
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Veri Kümesine Erişim
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
df = pd.read_csv(url, header=None, names=["CekirdekUzunlugu", "CekirdekGenisligi", "TaçUzunlugu", "TaçGenisligi", "Tur"])

# Veri Kümesini Görüntüleyin
print(df.head())
print(df.info())
print(df.describe())

# Veri Temizleme
# Araştırılacak çok fazla eksik değer olmadığında sadece kategorik değişken olan 'Tur' ile ilgileniyoruz.
print(df.isnull().sum())

# Keşifsel Veri Analizi (EDA)
# Türlere göre çekirdek uzunluğunu ve genişliğini görselleştirme
sns.scatterplot(x='CekirdekUzunlugu', y='CekirdekGenisligi', hue='Tur', data=df)
plt.title('Cekirdek Uzunlugu ve Genisligi')
plt.xlabel('Cekirdek Uzunlugu (cm)')
plt.ylabel('Cekirdek Genisligi (cm)')
plt.legend(title='Turler')
plt.show()

# Türlere göre taç uzunluğunu ve genişliğini görselleştirme
sns.scatterplot(x='TaçUzunlugu', y='TaçGenisligi', hue='Tur', data=df)
plt.title('Tac Uzunlugu ve Genisligi')
plt.xlabel('Tac Uzunlugu (cm)')
plt.ylabel('Tac Genisligi (cm)')
plt.legend(title='Turler')
plt.show()

# Kutu grafiği ile her bir özellik için türlere göre anlamlı farkları görselleştirme
plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
sns.boxplot(x='Tur', y='CekirdekUzunlugu', data=df)
plt.title('Cekirdek Uzunlugu')

plt.subplot(2, 2, 2)
sns.boxplot(x='Tur', y='CekirdekGenisligi', data=df)
plt.title('Cekirdek Genisligi')

plt.subplot(2, 2, 3)
sns.boxplot(x='Tur', y='TaçUzunlugu', data=df)
plt.title('Tac Uzunlugu')

plt.subplot(2, 2, 4)
sns.boxplot(x='Tur', y='TaçGenisligi', data=df)
plt.title('Tac Genisligi')

plt.tight_layout()
plt.show()

# Veri çerçevesinde bazı temel istatistiklerin analizi
tur_sayilari = df['Tur'].value_counts()
print("Tur Dagilimi:")
print(tur_sayilari)
```
